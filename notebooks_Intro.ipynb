```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from math import floor
```

```python
data=pd.read_csv('cost_revenue_clean.csv')
x=pd.DataFrame(data, columns=['production_budget_usd'])
y=pd.DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.title('Production Revenue VS Box Office Collection')
plt.xlabel('Production Budget in $')
plt.ylabel('Global Revenue in $')
plt.show()
```

```python
reg=LinearRegression()
reg.fit(x,y)
print("Revenue of Test Data = ","{:.2f}".format(reg.coef_[0][0])," * Movie_Budget")
```
